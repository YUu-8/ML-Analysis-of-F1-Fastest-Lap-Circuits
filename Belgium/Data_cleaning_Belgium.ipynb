### Data cleaning

```python
import pandas as pd
```

```python
df_import=pd.read_csv("lap_times_2025_round_13.csv")

df_import.head()
```

```python
nb_driver=df_import["Driver"].nunique()
print(nb_driver)
```

```python
#print(df['Lap Time'].dtype)

df_import["Lap Time"] = pd.to_timedelta(df_import["Lap Time"])
index = df_import.groupby("Driver")["Lap Time"].idxmin()
df=df_import.loc[index].reset_index(drop=True)

df.head()
```

```python
nb_driver_after=df["Driver"].nunique()
print(nb_driver_after)
```

```python
df["Lap Time"].isnull().sum() #v√©rifie qu'il n'y ait pas de valeur nulle
```

```python
list_column=["Driver","Lap Time","Sector 1","Sector 2","Sector 3"]
df=df[list_column]
df.head()
```

```python
print(df["Lap Time"].max())
```

```python
df["Lap Time"] = df["Lap Time"].astype(str).str.replace("0 days 00:", "", regex=False) #keep only minutes, seconds and hundredths of seconds
df["Sector 1"]=df["Sector 1"].astype(str).str.replace("0 days 00:", "", regex=False)
df["Sector 2"]=df["Sector 1"].astype(str).str.replace("0 days 00:", "", regex=False)
df["Sector 3"]=df["Sector 1"].astype(str).str.replace("0 days 00:", "", regex=False)

df.head()
```

```python
print("Number of none value for Sector 1 :" ,df["Sector 1"].isnull().sum())
print("Number of none value for Sector 2 :" ,df["Sector 2"].isnull().sum())
print("Number of none value for Sector 3 :" ,df["Sector 3"].isnull().sum())
```
